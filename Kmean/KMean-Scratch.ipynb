```python
def dis(red,blue):    #this function will calculate the mean/centroid of the clusters
    
    rmean=red.mean()    #mean of first cluster
    bmean=blue.mean()     #mean of second cluster
    x,y=initial(rmean,bmean)     #it will call another fuction to apply KNN
    return (x,y)     #it will return new clusters

def initial(rmean,bmean):
    a1=rmean.loc["X"]    #it will store the X value of 1st cluster
    b1=rmean.loc["Y"]    #it will store the Y value of 1st cluster
    a2=bmean.loc["X"]    #it will store the X value of 2nd cluster
    b2=bmean.loc["Y"]    #it will store the Y value of 2nd cluster
    red = pd.DataFrame({'X' : [np.nan],'Y' : [np.nan]})       #this will store all points of cluster 1
    blue = pd.DataFrame({'X' : [np.nan],'Y' : [np.nan]})      #this will store all points of cluster 2
    for i in range(len(df)):
        dfx=df.loc[i,"X"]
        dfy=df.loc[i,"Y"]
        #print(dfx,dfy)
        dis1=math.sqrt(math.pow(a1 - dfx, 2) + math.pow(b1 - dfy, 2) * 1.0)    #it will calculate the distamce from point to cluster 1
        
        dis2=math.sqrt(math.pow(a2 - dfx, 2) + math.pow(b2 - dfy, 2) * 1.0)    #it will calculate the distamce from point to cluster 2
        if (dis1>dis2):
            red=red.append(df.loc[i])    #if 1st distance is greater then it will add point to first cluster
        else:
            blue=blue.append(df.loc[i])  #else in the second cluster
        
    plt.scatter(red["X"],red["Y"])
    plt.scatter(blue["X"],blue["Y"])
    
    return (red,blue) #it will return the new clusters
```

```python
#selecting random points and applying knn
red1=df.loc[20]
blue1=df.loc[1500]
red,blue=initial(red1,blue1)
```
